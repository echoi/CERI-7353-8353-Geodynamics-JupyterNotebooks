# Question 2 <a name="question-2"></a>

The heat source term is inferred from the given condition as
\begin{align*}
F(z) = -\frac{H_{0}}{b} (z-b)
\end{align*}

Then, the steady-state heat equation with this source term becomes
\begin{align*}
0 = k\frac{\partial^{2}T}{\partial z^{2}} - \rho\,c_{p} F(z) = k\frac{\partial^{2}T}{\partial z^{2}} - \rho\,c_{p}\frac{H_{0}}{b}(z-b)
\end{align*}

Integrating twice with respecti to z, we get the following solution for temperature and heat flux with coefficients to be determined:
\begin{align*}
T(z) &= \frac{\rho c_{p} H_{0}}{b\,k} \left( \frac{1}{6} z^{3} - \frac{b}{2} z^{2} + c_{1} z + c_{2} \right) \\
q(z) &= -k\frac{\partial T}{\partial z} = -\frac{\rho c_{p} H_{0}}{b} \left( \frac{1}{2} z^{2} - b z + c_{1} \right)
\end{align*}

Applying boundary conditions, we determine $c_{1}$ and $c_{2}$.
Since the heat flux at $z=b$ is $-q_{m}$,

\begin{align*}
q(b) &= -q_{m} = -\frac{\rho c_{p} H_{0}}{b} \left( -\frac{1}{2} b^{2} + c_{1} \right) \\
\therefore \, c_{1} &= \frac{b\,q_{m}}{\rho c_{p} H_{0}} + \frac{b^{2}}{2}
\end{align*}

The surface temperature is $T_{0}$. So,
\begin{align*}
T(z=0) &= T_{0} = \frac{\rho c_{p} H_{0}}{b\,k}\,c_{2} \\
\therefore \, c_{2} &= \frac{b\,k\,T_{0}}{\rho c_{p} H_{0}}
\end{align*}

So, the temperature profile is
\begin{align*}
T(z) = T_{0} + \frac{q_{m}}{k}\,z + \frac{\rho c_{p} F_{s}}{b\,k} \left( \frac{1}{6}z^{3} - \frac{b}{2}z^{2} + \frac{b^{2}}{2}z \right),
\end{align*}
and the surface heat flux is
\begin{align*}
q_{s} = q(0) = q_{m} + \frac{1}{2}\rho\,c_{p}\,F_{s}\,b.
\end{align*}



# Question 3 <a name="question-3"></a>

Lord Kelvin assumed the instantaneous cooling half-space model:

\begin{equation}
  T(z) = T_{s} + (T_{m}-T_{s}) \, \text{erf} \left( \frac{z}{s\sqrt{\kappa\,t}} \right).
\end{equation}

In this model, the surface heat flow is given as

\begin{equation}
\begin{split}
  q_{s} &= - \left. k \frac{dT}{dz} \right|_{z=0} \\
        &= - k \, (T_{m}-T_{s}) \frac{1}{2\sqrt{\kappa\,t}} \frac{2}{\sqrt{\pi}} \exp\left[-\left(\frac{z}{2\sqrt{\kappa\,t}}\right)^{2} \right]_{z=0} \\
        &= - \frac{k \, (T_{m}-T_{s})}{\sqrt{\pi\,\kappa\,t}}.
\end{split}
\end{equation}

Using $q = 70 \ \text{mW/m}^{2}$, $\kappa = 10^{-6} \ \text{m}^{2}\text{s}^{-1}$, $k=2.5\ \text{W/m/K}$, and $T_{m}-T_{s} = 2000 \ \text{or} \ 4000 \ \text{K}$,
we get

\begin{equation}
  t = \frac{(T_{m}-T_{s})^{2}\,k^{2}}{\pi\,\kappa\,q^{2}} \approx 51 \ \text{or}\  206 \ \text{My}.
\end{equation}

The geochronologically measured age is much greater than these estimates. In other words, the Earth could somehow stay hot for a much longer time than Lord Kelvin thought. The discrepancy main comes from the fact that he did not consider the radioactive heating. Also, the advective heat transfer significantly contributes to the surface heat flow but was not considered.

## (a)

Under the stated assumptions, the heat advection-diffusion equation becomes

\begin{equation}
  \rho \, c_{p} \, v \, \frac{\partial T}{\partial x} = k \left( \frac{\partial^{2}T}{\partial x^{2}} + \frac{\partial^{2}T}{\partial z^{2}} \right).
\end{equation}

Defining non-dimensional temperature $T^{\prime}$ as $T = (T_{1}-T_{0})\,T^{\prime}$ and $x$ as $x = L \, x^{\prime}$, we get

\begin{align}
  \frac{\partial T}{\partial x} &= (T_{1}-T_{0}) \frac{\partial T^{\prime}}{\partial x}, \\
  \frac{\partial^{2} T}{\partial x^{2}} &= (T_{1}-T_{0}) \frac{\partial^{2} T^{\prime}}{\partial x^{2}}, \\
  \frac{\partial^{2} T}{\partial z^{2}} &= (T_{1}-T_{0}) \frac{\partial^{2} T^{\prime}}{\partial z^{2}}, \\  
\end{align}

Noting that 

\begin{align}
  \frac{\partial T^{\prime}}{\partial x^{\prime}} &= \frac{\partial T^{\prime}}{\partial x} \frac{\partial x}{\partial x^{\prime}} = L \frac{\partial T^{\prime}}{\partial x}, \\
  \frac{\partial^{2} T^{\prime}}{\partial x^{\prime\,2}} &= L^{2} \frac{\partial^{2} T^{\prime}}{\partial x^{2}}, \\
  \frac{\partial^{2} T^{\prime}}{\partial z^{\prime\,2}} &= L^{2} \frac{\partial^{2} T^{\prime}}{\partial z^{2}},  
\end{align}

we get the following equation in terms of the non-dimensional variables:

\begin{equation}
  \frac{\rho \, c_{p} \, v}{L} (T_{1}-T_{0}) \, \frac{\partial T^{\prime}}{\partial x^{\prime}} = \frac{k(T_{1}-T_{0})}{L^{2}} \left( \frac{\partial^{2} T^{\prime}}{\partial x^{\prime\,2}} + \frac{\partial^{2} T^{\prime}}{\partial z^{\prime\,2}} \right).
\end{equation}

After further simplification,

\begin{equation}
\begin{split}
  \frac{\partial T^{\prime}}{\partial x^{\prime}} &= \frac{k}{\rho \, c_{p} \, v \, L} \, \left( \frac{\partial^{2} T^{\prime}}{\partial x^{\prime\,2}} + \frac{\partial^{2} T^{\prime}}{\partial z^{\prime\,2}} \right) \\
  &= \frac{1}{2R} \, \left( \frac{\partial^{2} T^{\prime}}{\partial x^{\prime\,2}} + \frac{\partial^{2} T^{\prime}}{\partial z^{\prime\,2}} \right).
\end{split}
\end{equation}


## (b)

Now let's substitute the given solution in to the above equation:

\begin{equation}
\sum_{n=1}^{\infty} \left( -\alpha_{n}\,A_{n}\,e^{-\alpha\,x^{\prime}} \sin (\lambda_{n}\,z^{\prime}) \right)
=
\frac{1}{2R} \sum_{n=1}^{\infty} \left[ -\lambda_{n}^{2} \, A_{n} \, e^{-\alpha\,x^{\prime}} \sin (\lambda_{n}\,z^{\prime}) + \alpha_{n}^{2} \, A_{n} \, e^{-\alpha\,x^{\prime}} \sin (\lambda_{n}\,z^{\prime}) \right].
\end{equation}

Gathering the terms on one side, we get

\begin{equation}
\sum_{n=1}^{\infty} \left( -\alpha_{n} + \frac{\lambda_{n}^{2}-\alpha_{n}^{2}}{2R} \right) \, A_{n} \, e^{-\alpha\,x^{\prime}} \sin (\lambda_{n}\,z^{\prime}) = 0.
\end{equation}

To make the above euqation always hold,

\begin{equation}
 \alpha_{n} = \frac{\lambda_{n}^{2}-\alpha_{n}^{2}}{2R},
\end{equation}

or

\begin{equation}
 \alpha_{n}^{2} + 2\,R\,\alpha_{n} - \lambda_{n}^{2} = 0.
\end{equation}

\begin{equation}
 \therefore \ \alpha_{n} = -R + \sqrt{R^{2}+\lambda_{n}^{2}}.
\end{equation}


From the boundary condition that $T^{\prime}=0$ at $z^{\prime}=1$,

\begin{equation}
 T^{\prime}(z^{\prime}=1)=0 = 1 - 1 + \sum_{n=1}^{\infty} A_{n}\,e^{-\alpha\,x^{\prime}} \sin (\lambda_{n}),
\end{equation}

and this should be true for every $x^{\prime}$. Requiring that $A_{n} \neq 0$ for a non-trivial solution, we conclude that

\begin{equation}
 \lambda_{n} = n\,\pi.
\end{equation}

From another boundary condition that $T^{\prime}=1$ at $x^{\prime}=0$ and $0 \le z^{\prime} < 1$,

\begin{equation}
 T^{\prime}(x^{\prime}=0)= 1 = 1 - z^{\prime} + \sum_{n=1}^{\infty} A_{n} \, \sin (\lambda_{n}\,z^{\prime}),
\end{equation}

or 

\begin{equation}
 \sum_{n=1}^{\infty} A_{n} \, \sin (n\,\pi\,z^{\prime}) = z^{\prime}.
\end{equation}

The above relation should hold even after both sides are multiplied with $\sin (m\,\pi\,z^{\prime})$ and integrated over $0 \le z^{\prime} \le 1$:

\begin{equation}
 \int_{0}^{1} \sum_{n=1}^{\infty} A_{n} \, \sin (n\,\pi\,z^{\prime}) \, \sin (m\,\pi\,z^{\prime}) \, dz^{\prime} = \int_{0}^{1} z^{\prime} \, \sin (m\,\pi\,z^{\prime}) \, dz^{\prime}.
\end{equation}

Since the left hand side is non-zero only when $m=n$,

\begin{equation}
  A_{n} \, \int_{0}^{1} \sin^{2} (n\,\pi\,z^{\prime}) \, dz^{\prime} = \int_{0}^{1} z^{\prime} \, \sin (n\,\pi\,z^{\prime}) \, dz^{\prime}.
\end{equation}

Evaulating the integrals, we get

\begin{equation}
  A_{n} = \frac{2(-1)^{n+1}}{n\pi}.
\end{equation}


## (c)

To make the time dependence of $T$ explicit, we replace $x^{\prime}$ with the following expression:

\begin{equation}
x = v\,t \quad \rightarrow \quad x^{\prime} = \frac{x}{L} = \frac{v}{L}\,t.
\end{equation}

Combining all the results derived above, the temperature solutions is given as

\begin{equation}
  T^{\prime} = 1 - z^{\prime} + \sum_{n=1}^{\infty} \frac{2(-1)^{n+1}}{n\pi} \, \exp\left[ -\left(\sqrt{R^{2}+n^{2}\pi^{2}}-R \right) \frac{v}{L}t \right] \, \sin (n\,\pi\,z^{\prime}).
\end{equation}

The surface heat flow is then

\begin{equation}
\begin{split}
  q_{s}(t) &= - \left. k \frac{\partial T}{\partial z} \right|_{z=L} = -k \frac{T_{1}-T_{0}}{L} \left. \frac{\partial T^{\prime}}{\partial z^{\prime}} \right|_{z^{\prime}=1} \\
  &= -k \frac{T_{1}-T_{0}}{L} \left[ -1 + \sum_{n=1}^{\infty} 2(-1)^{n+1} \, \exp\left[ -\left(\sqrt{R^{2}+n^{2}\pi^{2}}-R \right) \frac{v}{L}t \right] \, \cos (n\,\pi) \right] \\
  &= k \frac{T_{1}-T_{0}}{L} \left[ 1 + 2 \, \sum_{n=1}^{\infty} \exp\left[ -\left(\sqrt{R^{2}+n^{2}\pi^{2}}-R \right) \frac{v}{L}t \right] \right].
\end{split}
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# constants
v = 5.0 * 0.01 / 3.1536e7 # 5 cm/yr to 5 m/sec.
My = 3.1536e13
rho = 3300.0 # kg/m^3
cp = 1000.0  # J/kg/K
k = 3.0      #
T1 = 1539.0  # deg C
T0 = 0.0     # deg C
rho_cp_v_over_k = rho * cp * v / k
```

```python
#
# Data and arrays
#

# Unsorted age and heat flow from the tables.
time = np.array([3.5,5.25,55.0,15.0,7.5,17.5,85.0,95.0,130.0,170.0,180.0,165.0,146.0,130.0,119.0,149.5,160.0,165.0]) * My
qdata0 = np.array([6.40,5.72,0.92,2.85,4.1,2.4,1.36,1.57,1.13])*41.84
qdata = np.append(qdata0, np.array([46.8,49.4,44.2,49.5,47.0,47.0,49.0,47.0,49.0]))

# Sort
ind = np.lexsort((qdata,time))
sorted_time = np.array([ time[i] for i in ind ])
sorted_qdata = np.array([ qdata[i] for i in ind ])

# z' array
zprime = np.linspace(0.0, 1.0, 101)
```

```python
def compute_Tprime(zprime, t, L):
    R = 0.5 * rho * cp * v * L / k
    v_over_L = v/L
    sum = 0.0
    for n in range(1,5001):
        coeff = (np.sqrt(R**2+(n*np.pi)**2)-R) * v_over_L
        sum = sum + (-1.0)**(n+1.0)/(n*np.pi) * np.exp(-coeff*t) * np.sin(n*np.pi*zprime)
    return 1.0 - zprime + 2.0 * sum

def compute_heatflow(t, L):
    R = 0.5 * rho * cp * v * L / k
    v_over_L = v/L
    q = np.zeros(len(t))
    for i in range(len(t)):
        sum = 0.0
        time = t[i]
        for n in range(1,501):
            coeff = (np.sqrt(R**2+(n*np.pi)**2)-R) * v_over_L
            sum = sum + np.exp(-coeff*time)
        q[i] = k * (T1 - T0) / L * (1.0 + 2.0 * sum)
    return q
```

```python
# Compute temperature profiles at various ages for a 100 km-thick plate.
Temp0 = compute_Tprime(zprime,0.1*My,100.0e3)
Temp1 = compute_Tprime(zprime,1.0*My,100.0e3)
Temp2 = compute_Tprime(zprime,10.0*My,100.0e3)
Temp3 = compute_Tprime(zprime,50.0*My,100.0e3)
Temp4 = compute_Tprime(zprime,200.0*My,100.0e3)
```

```python
# compute heat flows at the times of measurements for various plate thicknesses.
qmodel95 = compute_heatflow(sorted_time, 80.0e3)
qmodel98 = compute_heatflow(sorted_time, 100.0e3)
qmodel100 = compute_heatflow(sorted_time, 120.0e3)
```

# Question 5 <a name="question-5"></a>

## (a)

After the erosion of a thickness $l$ of ground, a surface at which $T=T_{0}+\beta\,l$ would be exposed on the surface. Since the surface temperature is maintained at $T_{0}$, the appropriate half-space cooling model would be

\begin{equation}
  \frac{T-T_{m}}{T_{0}-T_{m}} = \text{erfc} \left( \frac{y}{2\sqrt{\kappa\,t}} \right) = 1 - \text{erf} \left( \frac{y}{2\sqrt{\kappa\,t}} \right),
\end{equation}

where $T_{m} = T_{0}+\beta\,l$. 

From this, we get the expression for $T(y,t)$:

\begin{equation}
  T(y,t) = T_{0} + \beta\,l\,\text{erf} \left( \frac{y}{2\sqrt{\kappa\,t}} \right).
\end{equation}

The superposition of solutions is possible for linear PDEs like the linear heat diffusion equation. So, the contribution from the geothermal gradient, $\beta\,y$ can be simply added to the above solution:

\begin{equation}
  T(y,t) = T_{0} + \beta\,y + \beta\,l\,\text{erf} \left( \frac{y}{2\sqrt{\kappa\,t}} \right).
\end{equation}


## (b)

Before the erosion, the surface heat flow was just $-k\,\beta$.

After the erosion, the surface heat flow is given as
\begin{equation}
  q_{g}(y,t) = -k \left[ \beta + \frac{\beta\,l}{\sqrt{\pi\,\kappa\,t}} \right].
\end{equation}

Note that the assumption of the instanteneous erosion make the heat flow singular at $t=0$. Afterwards, however, the initially high heat flow falls back towards the background value of $-k\,\beta$ as $1/\sqrt{t}$.

# Question 6

## (a)

### For $t \le -\tau$:

\begin{equation}
  T = T_{0} + \beta\,y.
\end{equation}

\begin{equation}
  q = -k\,\beta.
\end{equation}

### For $-\tau < t \le 0$:

The surface temperature $T_{0}$ is reduced by $\Delta T$. Adopting the consideration in the previous question, we first compute the contribution from the half-space cooling with $T=T_{0}$ and the boundary temperature of $T_{0}-\Delta T$:

\begin{equation}
  \frac{T-T_{0}}{(T_{0}-\Delta T)-T_{0}} = \text{erfc} \left( \frac{y}{2\sqrt{\kappa\,(t+\tau)}} \right) = 1 - \text{erf} \left( \frac{y}{2\sqrt{\kappa\,(t+\tau)}} \right),
\end{equation}

Superposing the background geotherm on the above, we get the time-dependent temperature distribution during the glacial epoch ($T_{g}$):

\begin{equation}
  T_{g}(y,t) = (T_{0}-\Delta T) + \beta\,y + \Delta T\,\text{erf} \left( \frac{y}{2\sqrt{\kappa\,(t+\tau)}} \right).
\end{equation}

The heat flow during this epoch is 

\begin{equation}
  q_{g}(y,t) = -k \left[ \beta + \frac{\Delta T}{\sqrt{\pi\,\kappa\,(t+\tau)}} \exp \left( -\left( \frac{y}{2\sqrt{\kappa\,(t+\tau)}} \right)^{2}  \right) \right].
\end{equation}


### For $t > 0$:

The surface temperature has risen back to $T_{0}$. Then, the corresponding temperature change can be computed using the half-space cooling model again.
From

\begin{equation}
  \frac{T-(T_{0}-\Delta T)}{T_{0}-(T_{0}-\Delta T)} = \text{erfc} \left( \frac{y}{2\sqrt{\kappa\,t}} \right),
\end{equation}

\begin{equation}
  T(y,t) = T_{0} - \Delta T\,\text{erf} \left( \frac{y}{2\sqrt{\kappa\,t}} \right).
\end{equation}

Superposing the depth-dependent part of $T_{g}(t=0)$ on the above, we get

\begin{equation}
  T(y,t) = T_{0} - \Delta T\,\text{erf} \left( \frac{y}{2\sqrt{\kappa\,t}} \right) + \beta\,y + \Delta T\,\text{erf} \left( \frac{y}{2\sqrt{\kappa\,\tau}} \right).
\end{equation}

The heat flow is given as

\begin{equation}
  q(y,t) = -k \left[ \beta + \frac{\Delta T}{\sqrt{\pi\,\kappa\,\tau}} \exp \left( -\left( \frac{y}{2\sqrt{\kappa\,\tau}} \right)^{2}  \right) - \frac{\Delta T}{\sqrt{\pi\,\kappa\,t}}  \exp \left( -\left( \frac{y}{2\sqrt{\kappa\,t}} \right)^{2}  \right) \right].
\end{equation}


## (b)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
from scipy.special import erf
```

```python
yr2sec = 365.24*24.0*3600.0
tau = 5e3 * yr2sec # 5000 yrs in sec
kappa = 1e-6       # m^2/s
k = 3.3            # W/m/K
T0 = 283.0         # K
DT = 20.0          # K
beta = 0.0015      # K/m

y  = np.linspace(0.0, 3.0e3, 101) # depth array
tb = np.linspace(-1.5*tau,-tau,101) # time before glacial epoch
tg = np.linspace(-tau,0.0,101)      # glacial epoch
tp = np.linspace(0.0,0.5*tau,101)   # post-glacial epoch
```

```python
def Tb(y):
    return T0 + beta * y
def qb(y):
    return -k*beta
def Tg(y,t):
    eta = 0.5*y/np.sqrt(kappa*(t+tau))
    return (T0-DT) + beta*y + DT*erf(eta)
def qg(y,t):
    DT1 = DT/np.sqrt(np.pi*kappa*(t+tau))
    eta = 0.5*y/np.sqrt(kappa*(t+tau))
    return -k*(beta+DT1*np.exp(-eta**2))
def Tp(y,t):
    eta = 0.5*y/np.sqrt(kappa*t)
    etatau = 0.5*y/np.sqrt(kappa*tau)
    return T0 - DT*erf(eta) + beta*y + DT*erf(etatau)
def qp(y,t):
    DT1 = DT/np.sqrt(np.pi*kappa*tau)
    DT2 = DT/np.sqrt(np.pi*kappa*t)
    eta = 0.5*y/np.sqrt(kappa*t)
    etatau = 0.5*y/np.sqrt(kappa*tau)
    return -k*(beta+DT1*np.exp(-etatau**2)-DT2*np.exp(-eta**2))
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(10,5))
plt.rcParams.update({'font.size': 12})

ax1.plot(Tb(y), -y, '-', linewidth=2, color='black',    label='Background')
ax1.plot(Tg(y,-0.75*tau), -y , '-', linewidth=2, color='blue', label='$-0.75 \tau$')
ax1.plot(Tg(y,-0.25*tau), -y , '-', linewidth=2, color='cyan', label='$-0.25 \tau$')
ax1.plot(Tp(y,0.25*tau), -y , '-', linewidth=2, color='orange', label='$0.25 \tau$')
ax1.plot(Tp(y,0.75*tau), -y , '-', linewidth=2, color='red', label='$0.75 \tau$')

ax1.set_xlabel('$T^{\prime}$')
ax1.set_ylabel('z\'')
ax1.legend(loc=3)

ax2.plot(tb,np.ones(len(tb))*qb(0.0), 'o', label='Measurements' )
ax2.plot(tg,qg(0.0,tg), 'o', label='Measurements' )
ax2.plot(tp,qp(0.0,tp), 'o', label='Measurements' )
ax2.legend()
```
